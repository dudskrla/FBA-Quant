### Problem 1. 


Let $ W(t),0≤t≤T $, be a Brownian motion relative to a filtration $F(t),0≤t≤T. $ Let $ X $ be an $ F(T)-measurable $ random variable. For $ 0≤t≤T $ , define $ M(t)=E[X∣F(t)] $ .


(a) Show that $ M(t),0≤t≤T,$ is a martingale.

(b) Compute $ M(t) $ for $ 0≤t≤T $ when $ X=e^(W(T)) $ .

(c) Compute $ M(t) $ for $ 0≤t≤T $ when $ X=W^2 (T) $.

(d) Compute $ M(t) $ for $ 0≤t≤T $ when $ X=W^3 (T) $ .

(e) Compute $ M(t) $ for $ 0≤t≤T $ when $ X=∫_0^T W^2 (u)du $ .


### Solution 1.

**(a) Show that $ M(t),0≤t≤T,$ is a martingale.**

Martingale : For any $ s < t $ we have $ E(M(t) | F_s) = M(s) $, 

$$ E[M(t)|F(s)] = E[E[X|F(t)]|F(s)] = E[X|F(s)] = M(s) \; (∵ \; tower \; property : E[E[X|F(t)]|F(s)] = E[X|F(s)]) $$ 

**(b) Compute $ M(t) $ for $ 0≤t≤T $ when $ X=e^{W(T)} $ .**

$$ E(e^{W_T}|F_t) = E(e^{W_T-W_t+W_t}|F_s) = e^{W_t}E(e^{W_T-W_t}|F_s) $$

Since $ W_T - W_t$ ~ $ N(0, T-t) $, $ E(e^{W_T-W_t}) $ can be considered as moment generating function for normal distribution ($  M_X(t) = e^{\mu t+\frac{1}{2}\sigma^{2}t^2} $)

$$ = e^{W_t}e^{0 + \frac{1}{2}(T-t)} = e^{W_t + \frac{1}{2}(T-t)} $$


**(c) Compute $ M(t) $ for $ 0≤t≤T $ when $ X=W^2 (T) $.**


$$ E(W(T)^2∣F_t ) = E((W_t + W_T-W_t)^2 | F_t) = E((W_t + (W_T-W_t)^2) | F_t) $$

$$ = E(W_t^2 + 2W_t(W_T-W_t)+(W_T-W_t)^2| F_t) $$

$$ = E(W_t^2|F_t) + 2*E(W_t|F_t) * E((W_T-W_t)|F_t) + E((W_T-W_t)^2|F_t)  $$

$$ = W_t^2 + 2W_tE(W_T-W_t | F_t)+ E((W_T-W_t)^2|F_t) $$

$$ = W_t^2 + 0 + E((W_T-W_t)^2|F_t) $$

$$ ∴ \; E(W(T)^2|F_t) = W_t^2 + T-t $$

**(d) Compute $ M(t) $ for $ 0≤t≤T $ when $ X=W^3 (T) $ .**


$$ E(W(T)^3∣F_t ) = E((W_t + W_T-W_t)^3 | F_t) = E((W_t + (W_T-W_t))^3 | F_t) $$

$$ = E(W_t^3 + 3W_t^2(W_T-W_t)+3W_t(W_T-W_t)^2+(W_T-W_t)^3| F_t) $$

$$ = E(W_t^3|F_t) + 3*E(W_t^2|F_t) * E((W_T-W_t)|F_t) + E(3W_t(W_T-W_t)^2|F_t) + E((W_T-W_t)^3|F_t)  $$

$$ = W_t^3 + 3W_t^2E(W_T-W_t | F_t)+ 3W_tE((W_T-W_t)^2|F_t) + E((W_T-W_t)^3|F_t) $$

$$ = W_t^3 + 0\ + 3W_t(T-t)+E((W_T-W_t)^3|F_t) $$

$$ =W_t^3 + 3W_t(T-t)+E((W_T-W_t)^3|F_t) $$

$$ ∴ \; E(W(T)^3|F_t) = W_t^3 + 3(T-t)W_t $$

**(e) Compute $ M(t) $ for $ 0≤t≤T $ when $ X=∫_0^T W^2 (u)du $ .**

$$ E(∫_0^T W^2 (u)du |F_t) = E(∫_0^t W^2 (u)du + ∫_t^T W^2 (u)du |F_t)  $$

$$ = E(∫_0^t W^2 (u)du|F_t) + E(∫_t^T W^2 (u)du |F_t) = ? $$

$$ df(t, X(T)) = ∂_tf(t, X(t))dt + ∂_xf(t, X(t))dX(t) + \frac{1}{2}∂^2_xf(t, X(t))d[X, X](t) $$
