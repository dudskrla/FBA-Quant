### No, salvage value is not always inversely proportional to the useful life. 
Consider two virtual companies, XYZ Manufacturing and SwiftTech Electronics, to illustrate this point:

1) XYZ Manufacturing
XYZ Manufacturing produces heavy machinery for construction. These machines have a longer useful life due to the durability required in the construction industry. However, the salvage value might not be proportionally low. While the machinery lasts a long time, technological advancements or shifts in industry practices could lead to a relatively lower salvage value.

2) SwiftTech Electronics
SwiftTech Electronics manufactures consumer electronics, which have shorter lifespans due to rapid technological changes. Despite the short useful life, the salvage value might not necessarily be higher. The rapidly evolving technology landscape can cause even recent products to lose significant value in a short time.

In these examples, it's clear that salvage value and useful life are influenced by factors beyond a direct inversely proportional relationship. Technological trends, market demand, and maintenance practices all play roles in determining salvage value and useful life.

## [S2]

### 1) Purchase of Laboratory Equipment
In this scenario, BioMed Innovations acquires state-of-the-art laboratory equipment for $2,500,000 in cash during December. The recognized expense is indicated as $0. The key reason behind recognizing an expense of $0 in this case is due to the application of the accounting principle known as the "Matching Principle." According to this principle, expenses are recognized when they contribute to generating revenue.

Since the laboratory equipment was purchased and not yet put into use or sold during the same accounting period (December), it doesn't directly contribute to generating revenue. Therefore, in accordance with the Matching Principle, the expense associated with the laboratory equipment is not recognized immediately. Instead, the cost of the equipment is capitalized on the company's balance sheet as an asset until the equipment is used or sold in future periods. 

### 2) Sale of Medical Devices + Purchase of Laboratory Equipment
In this scenario, BioMed Innovations sells medical devices that were initially purchased for $7,000,000, achieving a total sales revenue of $12,000,000. The recognized expense is stated as $7,000,000. The reason for recognizing an expense of $7,000,000 here is due to the principle of expense recognition associated with the revenue generated.

When BioMed Innovations sells the medical devices, the cost associated with these devices, which was originally $7,000,000, is recognized as an expense. This expense is referred to as the "cost of goods sold." The cost of goods sold represents the direct costs incurred to produce the goods being sold, and it is a fundamental component in determining the company's gross profit margin. 

In summary, the difference in expense recognition between the two scenarios stems from the underlying principles of matching revenues with expenses.

## [S3]

### 1) Calculating Equity
- Equity = Total Assets - Total Liabilities    
- Equity = $1,200,000 - $500,000 = $700,000

### 2) Calculating Retained Earnings
- Retained Earnings = Prior Retained Earnings + Net Income - Dividends    
    - Net Income = Revenues - Expenses    
    - Net Income = $800,000 - $400,000 = $400,000    
- Retained Earnings = $100,000 + $400,000 - $50,000 = $450,000    

### 3) Calculating Contributed Capital
- Contributed Capital = Equity - Retained Earnings    
- Contributed Capital = $700,000 - $450,000 = $250,000

### 3) FRED

```python
# !pip install fredapi

# 1) Import libraries 
from fredapi import Fred
import pandas as pd

# 2) Authenticate using your personal API key
fred_api = "YOUR_API_KEY_HERE"
fred = Fred(api_key = fred_api)

# 3) Download the data
yield_curve = ["1MO", "3MO", "6MO", "1", "2", "3", "5", "7", "10", "20", "30"]
df = pd.DataFrame(columns = yield_curve)

for period in yield_curve:
    code = "DGS" + period
    
    try:
        yields = fred.get_series(code, sort_order="desc", limit=60, frequency="m")
    except Exception as e:
        print(f"Error occurred!: {e}") 
    else:
        df.loc[:, period] = yields

# 4) Inspect the downloaded data   
df = df.dropna()
df = df.astype(float)
df.index = pd.DatetimeIndex(df.index).date
df = df.sort_index()
display(df.head())
```
