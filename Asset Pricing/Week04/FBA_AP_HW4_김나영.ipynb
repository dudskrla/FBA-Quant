### Week 5. Introduction and Working Capital Assets

> Recognizing Uncollectible Accounts 
- **Direct write-off method**
    - Use : for tax reporting 
    - Method : recognize expense when account deemed uncollectible 
- **Allowance methods**
    - Use : for financial reporting
    - Method : recognize bad debt expense for estimated future uncollectible amounts from sales during the period

> Allowance method
$$ Net \; Accounts \; Receivable = (Gross) \; Accounts \; Receivable - Allowance \; for \; Doubtful \; Accounts $$ 

- **Percentage of sales method**
    - (direct) bad debt expense -> (indirect) ending balance of allowance for doubtful accounts 

    $$ Bad \; Debt \; Expense = Credit \; Sales * Estimated \; Uncollectible \; Percentage $$

- **Aging of accounts receivable method**
    - (direct) ending balance of allowance for doubtful accounts -> (indirect) bad debt expense

    $$ Ending \; Balance = \sum{Allowance \; Ammount} = \sum{Accounts \; Receivable * Estimated \; Uncollectible \; Percentage} $$

> Collecting cash from Accounts Receivable more quickly
- **Pledging**
    - Use A/R as collateral for a loan
- **Factoring**
    - Sell A/R to a financial institution (ex. bank) at a discount (that reflects an interest charge and the risk of uncollectibility)
- **Securitization**
    - Sell A/R to a seperate legal entity (called a Variable Interest Entity, VIE) 
        - The VIE borrows money from investors and then uses the proceeds to buy the A/R from its parent.

> Uncovering a change in Bad Debt Expense
- 1) Compute the percentage of Gross Accounts Receivable (that are expected to be uncollectible accounts in prior year)
    $$ Percent \; uncollectible = \frac{Allowance}{Net \; A/R + Allowance}$$ 
- 2) Apply this percentage to balance of Gross Accounts Receivable in current year to get expected balance 
    $$ Expected \; balance \; of \; Allowance = Percent \; uncollectible * (Net \; A/R + Allowance) $$
- 3) Change in bad debt assumptions
    $$ Change \; in \; bad \; debt \; expense = Actual \; balance - Expected \; balance $$

> Inventory Cost Flows for a Retail Firm VS. a Manufactoring Firm 



> Inventory and COGS 

$$ Beginning \; Inventory + New \; Inventory = COGS + Ending \; Inventory $$
- Ending Inventory = $ min(historical \; cost, \; fair \; market \; value) $ 
    - Historical cost 
        - the original cost of purchasing or producing the inventory
    - fair market value 
        - the replacement cost of the inventory (given the current market prices)

> Inventory cost flow assumptions
- FIFO (First In, First Out)
    - Oldest inventory costs fo in ***COGS*** first 
    - ***Ending inventory*** has newest costs 
- LIFO (Last In, First Out)
    - Newest inventory costs fo in ***COGS*** first 
    - ***Ending inventory*** has oldest costs 

> Comparing LIFO VS. FIFO firms
- Q. What is the FIFO value of the inventory?
    $$ FIFO \; End \; Inventory = LIFO \; End \; Inventory + LIFO \; Reserve $$ 
- Q. What would be FIFO COGS in this year?
    $$ FIFO \; COGS = LIFO \; COGS - ΔLIFO \; Reserve $$ 
- Q. What were tax savings in this year?
    $$ Tax \; savings = ΔLIFO \; Reserve * tax \; rate$$ 

### Week 6. Long-lived Assets and Marketable Securities

> Long-lived assets
- an asset that will provide benefits for **more than one year**

> Capitalizing VS. Expensing
- the timing of when the expense hits your income statement
    ex. Capitalizing (year 1 ~ year 10) VS. Expensing (year 1)

> Accounting issues about tangible assets
- 1) **Acquisition cost**    
    Capitalizing all costs necessary to get an asset ready for its intended use 
        - (1) increase the useful life of the asset
        - (2) increase the value of the asset
- 2) **Depreciation and amortization**    
        - (1) **Depreciable basis**
            $$ Acquisition \; cost - Salvage \; value $$
        - (2) **Useful life**
        - (3) **Depreciation pattern**  
            Total depreciation expense is the same over the lif of the asset (regardless of method used)
            - a) Straight line
                $$ \frac{Acquisition \; cost - Salvage \; value}{Useful \; life} $$
            - b) Accelerated  
- 3) **Ongoing costs**
    Expense costs related to the maintenanace of an asset
- 4) **Disposals**
    $$ Net \; book \; value = historical \; cost - accumulated \; depreciation $$
    - (1) all related **depreciation** is removed from accumulated depreciation
    - (2) **historical cost** of asset is removed from PP&E account
    - (3) *Gain or loss on the sale** of PP&E is recorded when the preceeds from disposal are more or less than the net book value
        - (a) Net Book Value > Cash → Loss on sale 
        - (b) Net Book Value < Cash → Gain on sale 
- 5) **Impairments**
    - (1) Dr. Equipment (+A) : the fair market value of the equipment
    - (2) Cr. Equipment (-A) : the prior value of the equipment 
    - (3) Dr. Accumulated Depreciation (-XA) 
    - (4) Dr. Impairment Loss (+E) : (to match Cr. == Dr.) 

> Accounting issues about intangible assets
- 1) Purchased Intangibles
    - added in an acquisition costs
    - (1) definite life → amortized over that life 
    - (2) indefinite life → not amortized 
- 2) Goodwill   
    - (1) the target's individual assets and liabilities are first adjusted to reflect their current market values
    - (2) the acquirer records any separable or transferable intangible assets 
    - (3) $ Goodwill = the \; excess \; of \; the \; purchase \; price -the \; market \; value \; of \; net \; identifiable \;  assets $

        - (a) Trading Securities

        - (b) Available for sale Securities

        - (c) Held to Maturity (debt)

A count of inventory at the end of the year revealed a balance in Finished Goods of $400,000 at 12/31/2012. What was Cost of Goods Sold during 2012?

## [Q2]

A company's customers owe it a total of $10,000 on 12/31/2012. On the same date, the company has an Allowance for Doubtful Accounts of $1,000. How much cash does the company expect to collect from customers that owe it money as of 12/31/2012?

## [Q3]

A company had credit sales of $500,000 during the quarter ended 9/30/2013. It had to write-off $300 of accounts as uncollectible during the quarter, and had no recoveries. Its balance in Allowance for Doubtful Accounts was $2,000 at 6/30/2013. Based on an aging of its accounts receivable, it expects that $2,700 of its outstanding A/R as of 9/30/2013 will ultimately be uncollectible in the future. How much Bad Debt Expense should the company report for the quarter ended 9/30/2013?

## [S1]

### \# Inventory \# COGS



- Beginning Inventory = $200,000
- New Inventory = $1,000,000
- Ending Inventory = $400,000

$$ COGS = Beginning \; Inventory + New \; Inventory - Ending \; Inventory $$
$$ = \$200,000 + \$1,000,000 - \$400,000 = \$800,000 $$ 


## [S2]

### \# Allowance for Doubtful Account



- (Gross) Accounts Receivable : $10,000
- Allowance for Doubtful Accounts : $1,000

$$ Net \; Accounts \; Receivable = (Gross) \; Accounts \; Receivable - Allowance \; for \; Doubtful \; Accounts \; (XA) $$
$$ = \$10,000 - \$1,000 = \$9,000 $$

## [S3]

### \# Bad Debt Expense

(1) Credit sales : A company had **credit sales** of $500,000 during the quarter ended 9/30/2013.    
(2) Write-offs : It had to **write-off** $300 of accounts as uncollectible during the quarter.    
(3) Allowance for Doubtful Accounts : Its balance in **Allowance for Doubtful Accounts** was $2,000 at 6/30/2013.    
(4) Accounts Receivable (time outstanding) : It expects that $2,700 of its outstanding **A/R** as of 9/30/2013 will ultimately be uncollectible in the future.    


> Table A : Accounts Receivable

|Accounts Receivable (A)|Debit|Credit|
|--|--|--|
|(1) Credit sales|$500,000||                
|(2) Write-offs||$300|             

         
> Table B : Sales 

|Sales|Debit|Credit|
|--|--|--|
|(1) Credit sales||$500,000|


> Table C : Allowance for Doubtful Accounts 

|Allowance for Doubtful Accounts|Debit|Credit|
|--|--|--|
|(3) Beginning balance||$2,000|
|Bad debt expense||$ ?|
|(2) Write-offs|$300||
|(4) Ending balance ||$2,700|


> Table D : Bad Debt Expense 

|Bad Debt Expense |Debit|Credit|
|--|--|--|
|Bad debt expense|$ ?||

$$ (In \; Table \; C.) \; End \; Balance = Debit - Credit $$
$$ \$2,700 = \$2,000 + Bad \; debt \; expense - \$300 $$ 
$$ ∴ \; Bad \; debt \; expense = \$2,700 - \$2,000 + \$300 = \$1,000 $$ 

- (More)
    - Why "(3) Allowance for Doubtful Accounts" is written as "(3) Beginning balance"?
        - Because, its balance in **Allowance for Doubtful Accounts** was $2,000 at ***"6/30/2013"*** (compared to 9/30/2013).
    - Why "(4) Accounts Receivable (time outstanding)" is written in Table C as "(4) Ending balance", not in Table A?
        - Because, its balance was ***already computed*** by the "aging of accounts receivable method".

### 1. Basic visualization of time series data

```python
# 1. Import the libraries
import pandas as pd
import numpy as np
import yfinance as yf

# 2. Download Microsoft’s stock prices from 2020 and calculate simple returns
df = yf.download("MSFT",
    start="2020-01-01",
    end="2020-12-31",
    auto_adjust=False,
    progress=False)

df["simple_rtn"] = df["Adj Close"].pct_change()
df = df.dropna() # NaNs introduced by calculating the percentage change

# 3. Plot the adjusted close prices
df["Adj Close"].plot(title="MSFT stock in 2020")
# df.plot.line(y="Adj Close", title="MSFT stock in 2020")
```

```python
# 4. Plot the adjusted close prices and simple returns in one plot
df[["Adj Close", "simple_rtn"]].plot(subplots=True, sharex=True, title="MSFT stock in 2020")
```

- (More) we can create a similar plot to the previous one using matplotlib's object-oriented interface

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 1, sharex=True)

df["Adj Close"].plot(ax=ax[0])
ax[0].set(title="MSFT time series",
          ylabel="Stock price ($)")
 
df["simple_rtn"].plot(ax=ax[1])
ax[1].set(ylabel="Return (%)")

plt.show()
```

### 3. Creating interactive visualizations

```python
# !pip install cufflinks
# !pip install pandas_bokeh

# 1. Import the libraries and initialize the notebook display
import pandas as pd
import yfinance as yf
import cufflinks as cf
from plotly.offline import iplot, init_notebook_mode
import plotly.express as px
import pandas_bokeh

cf.go_offline()
pandas_bokeh.output_notebook()

# 2. Download Microsoft’s stock prices from 2020 and calculate simple returns
df = yf.download("MSFT",
    start="2020-01-01",
    end="2020-12-31",
    auto_adjust=False,
    progress=False)

df["simple_rtn"] = df["Adj Close"].pct_change()
df = df.loc[:, ["Adj Close", "simple_rtn"]].dropna()
df = df.dropna()

# 3. Create the plot using cufflinks
df.iplot(subplots=True, shape=(2,1),
    shared_xaxes=True,
    title="MSFT time series")
```

- (More) Annotations are one of those techniques and we can easily add them to the plots generated with plotly (we can do so with other libraries as well).

```python
# 1. Import the libraries
from datetime import date

# 2. Define the annotations for the plotly plot
selected_date_1 = date(2020, 2, 19)
selected_date_2 = date(2020, 3, 23)

first_annotation = {
 "x": selected_date_1,
 "y": df.query(f"index == '{selected_date_1}'")["Adj Close"].squeeze(),
 "arrowhead": 5,
 "text": "COVID decline starting",
 "font": {"size": 15, "color": "red"},
}

second_annotation = {
 "x": selected_date_2,
 "y": df.query(f"index == '{selected_date_2}'")["Adj Close"].squeeze(),
 "arrowhead": 5,
 "text": "COVID recovery starting",
 "font": {"size": 15, "color": "green"},
 "ax": 150,
 "ay": 10
}

# 3. Update the layout of the plot and show it
fig.update_layout(
 {"annotations": [first_annotation, second_annotation]}
)

fig.show()
```

### 4. Creating a candlestick chart

- Data

- 1. To get the data for plotting

```python
# 1. Import the libraries
import pandas as pd
import yfinance as yf

# 2. Download the adjusted prices
df = yf.download("AAPL",
    start="2018-01-01",
    end="2018-12-31",
    progress=False,
    auto_adjust=True)
```

- 2. To create an interactive candlestick chart

```python
# 1. Import the libraries
import cufflinks as cf
from plotly.offline import iplot

cf.go_offline()

# 2. Create the candlestick chart using Twitter’s stock prices
qf = cf.QuantFig(
    df, title="Apple's Stock Price", 
    legend="top", 
    name="Apple's stock prices in 2018"
)

# 3. Add volume and moving averages to the figure
qf.add_volume()
qf.add_sma(periods=20, column="Close", color="red")
qf.add_ema(periods=20, color="green")

# 4. Display the plot
qf.iplot()
```

- (More) To create candlestick charts using pure plotly (in case you do not want to use a wrapper library such as cufflinks) and mplfinance, a standalone expansion to matplotlib dedicated to plotting financial data

```python
# !pip install mplfinance

# 1. Import the libraries
import plotly.graph_objects as go
import mplfinance as mpf

# 2. Create a candlestick chart using pure plotly
fig = go.Figure(
    data=
        go.Candlestick(
            x=df.index,
            open=df["Open"],
            high=df["High"],
            low=df["Low"],
            close=df["Close"]
        )
)


fig.update_layout(
    title="Apple's stock prices in 2018",
    yaxis_title="Price ($)"
)

fig.show()
```

```python
# 3. Create a candlestick chart using mplfinance
mpf.plot(
    df, 
    type="candle",
    mav=(10, 20),
    volume=True,
    style="yahoo",
    title="Apple's stock prices in 2018",
    figsize=(8, 4)
)
```
