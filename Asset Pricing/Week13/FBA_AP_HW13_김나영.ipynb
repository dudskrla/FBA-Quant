### Chapter6. Mean-Variance Portfolio Theory
Mean-variance analysis forms the basis for the important capital asset pricing model discussed in Chapter 7.

**6.1 Asset Return**
- Total return 
$$ Total \; return \; (R) = \frac{amount \; received \; (X_1)}{amount \; invested \; (X_0)} $$

- The rate of return 
$$ The \; rate \; of \; return \; (r) = \frac{amount \; received \; (X_1) - amount \; invested \; (X_0)}{amount \; invested \; (X_0)} $$

- Relationship of total return (R) and the rate of return (r)
$$ R = 1 + r $$

- Short sales        
= to sell an asset that you do not own 
    $$ Total \; return \; (R) = \frac{the \; outlay \; (-X_1)}{final \; receipt \; (-X_0)} $$

- Portfolio return 
portfolio (= master asset) consisted of n different asets 
$$ R = \sum_{i=1}^n w_iR_i $$
$$ r = \sum_{i=1}^n w_ir_i $$

**6.2 Random Variables**
- (1) a finite number of specific values, say, $ x_1, x_2, ... ,x_m $
- (2) each possible $ x_i $, there is a probability P
- (3) The $ p_i $'s satisfy $ \sum_{i=1}^m p_i = 1 $
- (4) $ p_i  0 $ 


- Expected value ($ E(x), \bar{x} $)       
= the average value obtained by regarding the probabilities as frequencies

- Variance 
$$ var(x) = E[(x-\bar{x})^2] $$
$$ = E(x^2) - 2E(x)\bar{x} + \bar{x}^2 $$
$$ = E(x^2) - \bar{x}^2 $$ 

- Covariance
$$ cov(x_1, x_2) = E[(x_1-\bar{x}_1)(x_2-\bar{x}_2)] $$
$$ = E(x_1x_2) - \bar{x}_1\bar{x}_2$$

- Correlation coefficient
$$ _{12} = \frac{\sigma_{12}}{\sigma_1\sigma_2} $$ 

- Variance of a sum
$$ var(x+y) = E[(x-\bar{x} + y - \bar{y})^2] $$
$$ = E[(x-\bar{x})^2] + 2E[(x-\bar{x})(y-\bar{y})] + E[(y-\bar{y})^2] $$ 
$$ = \sigma_x^2 + 2\sigma_{xy} + \sigma_y^2 $$

**6.3 Random Returns**             
the rate of return is usually uncertain
- Expected value
$$ E(r) = \bar{r} $$
- Variance
$$ E[(r-\bar{r})^2] = \sigma^2 $$

**6.4 Portfolio Mean and Variance**
- Mean return of a portfolio
$$ r = w_1r_1 + w_2r_2 + ... + w_nr_n $$
$$ E(r) = w_1E(r_1) + w_2E(r_2) + ... + w_nE(r_n) $$

- Variance of portfolio return 
$$ \sigma^2 = E[(r-\bar{r})^2] = \sum_{i, j = 1}^n w_iw_j\sigma_{ij} $$

- Diversification
$$ r = \frac{1}{n} \sum_{i=1}^n r_i $$
(1) individual returns are uncorrelated 
$$ var(r) = \frac{1}{n^2} \sum_{i=1}^n \sigma^2 = \frac{\sigma^2}{n} $$ 
(2) individual returns are correlated (assumption. $ cov(r_i, r_j) = \sigma^2$)
$$ var(r) = \frac{1}{n^2} \sum_{i=1}^n \sigma^2 = \frac{(1-) \sigma^2}{n} + \sigma^2 $$ 

- Diagram of a portfolio             

    - solid portion : positive combinations of the two assets 
    - dashed portion : shorting of one of them 

**6.5 The Feasible Set**   
= the set of points that correspond to portfolios               

- The minimum-variance set and the efficient frontier 
    - the minimum-variance set         
        the left boundary of a feasible set 

- risk averse 
the point with the smallest standard deiation for the given mean
- nonsatiation
the portfoilo of the largest mean for a given level of standard deviation

- efficient frontier
    - risk averse + nonsatiation 
    - upper portion of the minimum-variance set

**6.6 The Markowitz Model**
$$ minimize \; \frac{1}{2} \sum_{i, j = 1}^n w_iw_j\sigma_{ij} $$
$$ subject \; to \; \sum_{i=1}^n w_i\bar{r}_i = \bar{r} $$
$$ \sum_{i=1}^n w_i = 1 $$

- Solution of the Markowitz problem using Lagrange multipliers
$$ L=\frac{1}{2} \sum_{i,j=1}^2{w_i w_j \sigma_{ij}} - \lambda_1 (\sum_{i=1}^2{w_i _i-_0})-_2 (_{i=1}^2 {w_i-1}) $$ 
differentiate the Lagrangian with respect to each variable $ w_i $ and set this derivative to zero

- Equations for efficient set (n+2 linear equalities)
$$ \sum_{j=1}^n \sigma_{ij}w_j - \bar{r}_i -  = 0 \; for \; i=1, 2, ..., n $$
$$ \sum_{i=1}^n w_i\bar{r}_i = \bar{r} $$
$$ \sum_{i=1}^n w_i = 1 $$

- Equations for efficient set + nonnegativity constraints (n+2 linear equalities + n inequalities)
$$ \sum_{j=1}^n \sigma_{ij}w_j - \bar{r}_i -  = 0 \; for \; i=1, 2, ..., n $$
$$ \sum_{i=1}^n w_i\bar{r}_i = \bar{r} $$
$$ \sum_{i=1}^n w_i = 1 $$
$$ w_i  0 \; for \; i=1, 2, ..., n $$

**6.7 The Two-Fund Theorem**     
Two efficient funds (portfolios) can be established so that any efficient portfolio can be duplicated

There would be no need for anyone to purchase individual stocks separately; they could just purchase shares in the mutual funds.

**6.8 Inclusion of a Risk-Free Asset**
- Risk-free asset          
deterministic (known with certainty) and therefor has $=0$
$$ Mean = r_f + (1-)\bar{r} $$
$$ Standard \; deviation = (1-)\sigma $$ 

**6.9 The One-Fund Theorem**       

There is a single fund F of risky assets such that any efficient portfolio can be constructed as a combination of the fund F and the risk-free asset

- Solution
$$ tan = \frac{\bar{r}_P - r_f}{\sigma_P} $$
$$ = \frac{\sum_{i=1}^n w_i (\bar{r}_i - r_f) }{ (\sum_{i, j = 1}^n \sigma_{ij}w_iw_j)^{1/2} } $$
- (1) not necessary to impose the constraint $ \sum_{i=1}^n w_i = 1$
- (2) set the derivative of $ tan $ with the respect to the each $ w_k $ equal to zero 
    $$ \sum_{i=1}^n \sigma_{ki}w_i = \bar{r}_k - r_f, \; k=1, 2, ..., n $$
- (3) make substitution $ v_i =  w_i $ for each i
- (4) normalize to determine the $w_i$'s
    $$ w_i = \frac{v_i}{\sum_{k=1}^n v_k} $$

**3) CPPI and Drawdown Constraints-Part1**

```python
import numpy as np
import pandas as pd
import edhec_risk_kit as erk

ind_return = erk.get_ind_returns()
tmi_return = erk.get_total_market_index_returns()
```

```python
risky_r = ind_return["2000":][["Steel", "Fin", "Beer"]]

safe_r = pd.DataFrame().reindex_like(risky_r)
safe_r.values[:] = 0.03/12 
start = 1000 
floor = 0.8
```

```python
def compound1(r):
    return np.expm1(np.log1p(r).sum())

def compound2(r):
    return (r+1).prod()-1

compound1(risky_r)
```

```python
compound2(risky_r)
```

```python
dates = risky_r.index
n_steps = len(dates)
account_value = start
floor_value = start*floor
m = 3

account_history = pd.DataFrame().reindex_like(risky_r)
risky_w_history = pd.DataFrame().reindex_like(risky_r)
cushion_history = pd.DataFrame().reindex_like(risky_r)

for step in range(n_steps):
    cushion = (account_value - floor_value)/account_value
    risky_w = m*cushion
    risky_w = np.minimum(risky_w, 1)
    risky_w = np.maximum(risky_w, 0)
    safe_w = 1-risky_w
    risky_alloc = account_value*risky_w
    safe_alloc = account_value*safe_w
    
    account_value = risky_alloc*(1+risky_r.iloc[step]) + safe_alloc*(1+safe_r.iloc[step])

    cushion_history.iloc[step] = cushion
    risky_w_history.iloc[step] = risky_w
    account_history.iloc[step] = account_value
```

```python
risky_wealth = start*(1+risky_r).cumprod()
risky_wealth.plot()
```

```python
def show_gbm(n_scenarios, mu, sigma):
    """
    Draw the results of a stock price evolution under a Geometric Brownian Motion model
    """
    s_0=100
    prices = erk.gbm(n_scenarios=n_scenarios, mu=mu, sigma=sigma, s_0=s_0)
    ax = prices.plot(legend=False, color="indianred", alpha = 0.5, linewidth=2, figsize=(12,5))
    ax.axhline(y=100, ls=":", color="black")
    # draw a dot at the origin
    ax.plot(0,s_0, marker='o',color='darkred', alpha=0.2)
```

```python
gbm_controls = widgets.interactive(show_gbm, 
                                   n_scenarios=(1, 20, 5), 
                                   mu=(0., 0.2,0.01),
                                   sigma=(0, .3, .01)
)
```

```python
display(gbm_controls)
```

```python
def show_cppi(n_scenarios=50, mu=0.07, sigma=0.15, m=3, floor=0., riskfree_rate=0.03, y_max=100):
    """
    Plot the results of a Monte Carlo Simulation of CPPI
    """
    start = 100
    sim_rets = erk.gbm(n_scenarios=n_scenarios, mu=mu, sigma=sigma, prices=False, steps_per_year=12)
    risky_r = pd.DataFrame(sim_rets)
    # run the "back"-test
    btr = erk.run_cppi(risky_r=pd.DataFrame(risky_r),riskfree_rate=riskfree_rate,m=m, start=start, floor=floor)
    wealth = btr["Wealth"]
    y_max=wealth.values.max()*y_max/100
    ax = wealth.plot(legend=False, alpha=0.3, color="indianred", figsize=(12, 6))
    ax.axhline(y=start, ls=":", color="black")
    ax.axhline(y=start*floor, ls="--", color="red")
    ax.set_ylim(top=y_max)

cppi_controls = widgets.interactive(show_cppi, 
                                   n_scenarios=widgets.IntSlider(min=1, max=1000, step=5, value=50), 
                                   mu=(0., +.2, .01),
                                   sigma=(0, .30, .05),
                                   floor=(0, 2, .1),
                                   m=(1, 5, .5),
                                   riskfree_rate=(0, .05, .01),
                                   y_max=widgets.IntSlider(min=0, max=100, step=1, value=100,
                                                          description="Zoom Y Axis")
)
display(cppi_controls)
```


**8) interactive plots of monte Carlo Simulations of CPPI and GBM-Part2**

```python
import matplotlib.pyplot as plt

def show_cppi(n_scenarios=50, mu=0.07, sigma=0.15, m=3, floor=0., riskfree_rate=0.03, y_max=100):
    """
    Plot the results of a Monte Carlo Simulation of CPPI
    """
    start = 100
    sim_rets = erk.gbm(n_scenarios=n_scenarios, mu=mu, sigma=sigma, prices=False, steps_per_year=12)
    risky_r = pd.DataFrame(sim_rets)
    # run the "back"-test
    btr = erk.run_cppi(risky_r=pd.DataFrame(risky_r),riskfree_rate=riskfree_rate,m=m, start=start, floor=floor)
    wealth = btr["Wealth"]
    # calculate terminal wealth stats
    y_max=wealth.values.max()*y_max/100
    terminal_wealth = wealth.iloc[-1]
    # Plot!
    fig, (wealth_ax, hist_ax) = plt.subplots(nrows=1, ncols=2, sharey=True, gridspec_kw={'width_ratios':[3,2]}, figsize=(24, 9))
    plt.subplots_adjust(wspace=0.0)
    
    wealth.plot(ax=wealth_ax, legend=False, alpha=0.3, color="indianred")
    wealth_ax.axhline(y=start, ls=":", color="black")
    wealth_ax.axhline(y=start*floor, ls="--", color="red")
    wealth_ax.set_ylim(top=y_max)
    
    terminal_wealth.plot.hist(ax=hist_ax, bins=50, ec='w', fc='indianred', orientation='horizontal')
    hist_ax.axhline(y=start, ls=":", color="black")

cppi_controls = widgets.interactive(show_cppi, 
                                   n_scenarios=widgets.IntSlider(min=1, max=1000, step=5, value=50), 
                                   mu=(0., +.2, .01),
                                   sigma=(0, .3, .05),
                                   floor=(0, 2, .1),
                                   m=(1, 5, .5),
                                   riskfree_rate=(0, .05, .01),
                                   y_max=widgets.IntSlider(min=0, max=100, step=1, value=100,
                                                          description="Zoom Y Axis")
)
display(cppi_controls)
```

### Problem 3. 
Jins friend is planning to invest $1 million in a rock concert to be held 1 year from now. The friend figures that he will obtain $3 million revenue from his $1 million investment-unless, my goodness, it rains. If it rains, he will lose his entire investment. There is a 50% chance that it will rain the day of the concert. Gain suggests that he buy rain insurance. He can buy one unit of insurance for $0.5. and this unit pays $1 if it rains and nothing if it does not. He may purchase as many as he wishes, up to $3 million.

a) What is the expected rate of return on his investment if he buys u units of insurance? (The cost of insurance is in addition ot his $1 million investment.)

b) What number of units will minimize the variance of his return? What is this minimum value? And what is the corresponding expected rate of return?
Hint: Before calculating a general expression for variance, think about a simple answer.

### Solution 3.
**a) What is the expected rate of return on his investment if he buys u units of insurance? (The cost of insurance is in addition ot his $1 million investment.)**

$$ sol) \; \bar{R} = \frac{amounted \; received}{amounted \; invested} = \frac{p*(\$0 + \$1 * u) + (1-p)*(\$3*10^6 + \$0*u)}{\$10^6 + \$0.5*u} \; = \frac{\$0.5 * u+ \$1.5*10^6}{\$10^6 + \$0.5*u} $$
$$ u : units \; of \; insurance $$
$$ p : probability \; of \; raining \; (p=\frac{1}{2})$$
$$  \; \bar{r} = \bar{R}-1 = \frac{\$0.5 * u+ \$1.5*10^6}{\$10^6 + \$0.5*u} - 1 = \frac{\$0.5*10^6}{\$10^6 + \$0.5*u} $$ 

**b) What number of units will minimize the variance of his return? What is this minimum value? And what is the corresponding expected rate of return?
Hint: Before calculating a general expression for variance, think about a simple answer.**    

sol) Buying 3 million units of insurance removes any uncertainty in the amount received at the concert day. 
- **(1) What number of units will minimize the variance of his return?**        
    A. 3 million units
- **(2) What is this minimum value?**    
    A. 0 ( Whether it rains or not, expected revenue = $ \$3*10^6 $)
- **(3) What is the corresponding expected rate of return?**       
    A. $ \bar{r} = \frac{\$0.5*10^6}{\$10^6 + \$0.5*3*10^6} = 0.2 $
