### Problem 1. 
An investor deposits $15,000 in a bank today at 6% annual interest. Assuming that no additional deposits are made, and that no money is withdrawn, how large will the account balance be 23 years from today if    

(a) the interest is computed using the simple interest convention?   
(b) the interest is compounded annually?    
(c) the interest is compounded quarterly?    
(d) the interest is compounded monthly?    
(e) the interest is compounded continuously?    

(You should assume that the 6% annual interest rate applies to deposits of any length.)


> Simple interest convention

> Compound interest convention

â€‹


```python
import numpy as np

# (a) the interest is computed using the simple interest convention
a = 15000 * (1 + 0.06*23)
print(f"(a) = {a}")

# (b) the interest is compounded annually
b = 15000 * (1 + 0.06)**23
print(f"(b) = {b}")

# (c) the interest is compounded quarterly
c = 15000 * (1 + 0.06 / 4)**(23 * 4)
print(f"(c) = {c}")

# (d) the interest is compounded monthly
d = 15000 * (1 + 0.06 / 12)**(23 * 12)
print(f"(d) = {d}")

# (e) the interest is compounded continuously
e = 15000 * (np.exp(0.06*23))
print(f"(e) = {e}")
```

Suppose that the stockâ€™s price today is 100 and $\sigma^2$ = 1.

> Chebyshev inequality 

$$ P(|X - \mu| \geq k) \leq \frac{\sigma^2}{k^2} $$

#### (sol-4)

(a)
$$ E[S_{10}] = 100 + 10 * \mu = 100 $$

$$ Var(S_{10}) = 10 * \sigma ^ 2 = 10 $$

Therefore

$$ P(|S_{10} - 100| \geq k) \leq \frac{10}{k^2} $$

(b)
 $$ P(S_{10} \geq 105) = P (\frac{s_{10} - \mu}{\sigma} \geq \frac{105 - 100}{\sqrt{10}}) $$

```python
# (c)
import numpy as np
import matplotlib.pyplot as plt

M = 10000 
T = 10    
sqrt_sigma = 1

trajectories = np.zeros((M, T+1))
trajectories[:, 0] = 100  # Initial stock price

for i in range(M):
    for t in range(1, T+1):
        W_t = np.random.normal(0, np.sqrt(sqrt_sigma))
        trajectories[i, t] = trajectories[i, t-1] + W_t

# Plot 10 pairs of trajectories
plt.figure(figsize=(10, 6))
for i in range(10):
    plt.plot(range(T+1), trajectories[i], label=f'Trajectory {i+1}')

plt.xlabel('Days')
plt.ylabel('Stock Price')
plt.title('Stock Price Trajectories')
plt.legend()
plt.grid()

# Calculate the probability that the stock price exceeds 105 after 10 days
count_exceed_105 = np.sum(trajectories[:, -1] > 105)
probability_exceed_105 = count_exceed_105 / M

print("Estimated Probability:", probability_exceed_105)

plt.show()
```

, where all $ð‘Ž_ð‘–$ are positive and ð‘› > 1. Here is an iterative technique that generates a sequence $Î»_0, Î»_1, Î»_2, â€¦ , Î»_ð‘˜$, â€¦ of estimates that converges to the root $\bar{Î»}>0$, solving ð‘“($\bar{Î»}$) = 0. Start with any $Î»_0 > 0$ close to the

$$ ð‘“â€²(Î»_ð‘˜) = ð‘Ž_1 + 2ð‘Ž_2Î»_ð‘˜ + 3ð‘Ž_3Î»_ð‘˜^2 + â‹¯ + ð‘›ð‘Ž_ð‘›Î»_ð‘˜^{ð‘›âˆ’1} $$
and define
$$ Î»_{ð‘˜+1} = Î»_ð‘˜ âˆ’ \frac{ð‘“(Î»_ð‘˜)}{ð‘“â€²(Î»_ð‘˜)} $$ 

Try the procedure on $ð‘“(Î») = âˆ’1 + Î» + Î»^2$ starting with $Î»_0 = 1$.

(a) Write a Python function to calculate ðœ†Ì…accurate up to 0.000001 and compute the computation time    
(b) Use the Bisection method to calculate ðœ†Ì…accurate up to 0.000001 and compare the computation time with the result from (a)    


```python
import time

# (a)
def f(lambda_): # function f(Î»)
    return -1 + lambda_ + lambda_**2

def f_prime(lambda_): # Derivative of f(Î»)
    return 1 + 2*lambda_

def newton():
    gred = 1 # initial value (Î»_0 = 1)
    accr = 0.000001
    start_time = time.time()
    
    while abs(f(gred)) > accr:
        gred = gred - f(gred) / f_prime(gred)
        
    end_time = time.time()
    computation_time = end_time - start_time
    return gred, computation_time

answer_lambda_a, answer_time_a = newton()

print("Lambda:", answer_lambda_a)
print("Computation Time:", answer_time_a, "seconds\n")


# (b)
def bisect():
    start, end = 0, 1
    accr = 0.000001
    start_time = time.time()
    
    while (end - start) / 2 > accr:
        mid = (start + end) / 2
        if f(mid) == 0:
            return mid
        
        if f(start) * f(mid) < 0:
            end = mid
        else:
            start = mid
            
    end_time = time.time()
    computation_time = end_time - start_time
    return (start + end) / 2, computation_time

answer_lambda_b, answer_time_b = bisect()

print("Lambda:", answer_lambda_b)
print("Computation Time:", answer_time_b, "seconds\n")
```

### Problem 6. 
Given N points drawn randomly on the circumference of a circle, what is the probability that they are all within a semicircle

#### (sol - 6)
The probability that a single point falls within this semicircle is the ratio of the semicircle's area to the total circle's area.   
The area of a semicircle with radius $r$ is $\frac{1}{2}\pi r^2$, and the area of the whole circle is $\pi r^2$.     
So, the probability for one point is $\frac{\frac{1}{2}\pi r^2}{\pi r^2} = \frac{1}{2}$.

Since the points are randomly distributed, the probability that all $N-1$ remaining points fall within the semicircle is $\left(\frac{1}{2}\right)^{N-1}$, as the positions of the points are independent.

Therefore, the probability that all $N$ points fall within a single semicircle is $\left(\frac{1}{2}\right)^{N-1}$.
